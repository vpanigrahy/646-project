```python
import pickle
import pandas as pd
from rank_bm25 import BM25Okapi
```

```python
with open('restaurent_docs.pickle', 'rb') as f:
    docs_df = pickle.load(f)
```

```python
docs_df.head(7)
```

```python
documents = docs_df['doc_information'].to_list()
```

```python
document_ids = [i for i in range(len(documents))]
```

```python
tokenized_docs = [doc.split(" ") for doc in documents]
```

```python
bm25 = BM25Okapi(tokenized_docs)
```

```python
query = "Mobil Gas Station"
query_tokens = query.split(" ")

# Get document scores
doc_scores = bm25.get_scores(query_tokens)
```

```python
doc_scores_with_ids = list(zip(document_ids, doc_scores, docs_df['avg_rating'].to_list()))
```

```python
sorted_docs_with_ids = sorted(doc_scores_with_ids, key=lambda x: -x[1])
```

```python
ind = [i for i, _ , _ in sorted_docs_with_ids[:10]]
```
