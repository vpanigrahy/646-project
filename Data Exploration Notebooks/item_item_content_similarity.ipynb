```python
import pandas as pd
```

```python
restaurent_data = pd.read_pickle('restaurent_docs_1.pickle')
```

```python
restaurent_data
```

```python
restaurent_data = restaurent_data.drop_duplicates(subset='gmap_id', keep='first')
restaurent_data.reset_index(drop=True, inplace=True)
```

```python
restaurent_data
```

```python
def concat_dict_values(d):
    if d is None:
        return ''
    return ', '.join([f"{' '.join(value)}" for key, value in d.items()])


restaurent_data['similarity_text'] = restaurent_data['description'].fillna(' ') + ' ' + restaurent_data['category'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x)) + ' ' + restaurent_data['MISC'].apply(concat_dict_values)
```

```python
restaurent_data
```

```python
restaurent_data = restaurent_data[["gmap_id", "similarity_text"]]
```

```python
restaurent_data
```

```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def compute_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1)
    return embeddings

embeddings_dict = {}
for idx, row in restaurent_data.iterrows():
    gmap_id = row['gmap_id']
    similarity_text = row['similarity_text']
    embeddings_dict[gmap_id] = compute_embedding(similarity_text)
```

```python
similarity_matrix = pd.DataFrame(index=embeddings_dict.keys(), columns=embeddings_dict.keys())
for gmap_id1, emb1 in embeddings_dict.items():
    for gmap_id2, emb2 in embeddings_dict.items():
        similarity_score = cosine_similarity(emb1, emb2)[0][0]
        similarity_matrix.loc[gmap_id1, gmap_id2] = similarity_score
    
    print("gmap done : ",gmap_id1)

print(similarity_matrix)
```

```python
item_item_content_similarty = pd.DataFrame(similarity_matrix)
```
