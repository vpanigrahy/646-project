```python
import pandas as pd
import json

file_path = './meta-Texas.json'

data = []
with open(file_path, 'r') as file:
    for line in file:
        try:
            obj = json.loads(line)
            data.append(obj)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
            
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
restaurant_data = df[df['category'].apply(lambda x: 'restaurant' in str(x).lower() if x is not None else False)]

restaurant_data.shape
```

```python

restaurant_data.dropna(subset=['address'], inplace=True)
houston_rest_df = restaurant_data[restaurant_data['address'].str.contains('Houston', case=False)]
```

```python
houston_rest_df['price'].value_counts()
```

```python
houston_rest_df['price'].replace('₩', '$',regex=True, inplace=True)
```

```python
houston_rest_df['price'].value_counts()
```

```python
price_mapping = {None: 0, '$': 1, '$$': 2, '$$$': 3, '$$$$': 4}

houston_rest_df['price'] = houston_rest_df['price'].replace(price_mapping)
```

```python
houston_rest_df = houston_rest_df[houston_rest_df['state'] != 'Permanently closed']
```
