```python

def limit_words(text, target_word_count=10000):
    words = text.split()
    num_words_to_keep = min(len(words), target_word_count)
    limited_text = ' '.join(words[:num_words_to_keep])

    return limited_text

docs['doc_information'] = docs['doc_information'].apply(limit_words)

```

```python
docs
```

```python
from textstat import textstat

def count_words(text):
    return textstat.lexicon_count(text, removepunct=True)

docs['word_count'] = docs['doc_information'].apply(count_words)
```

```python
docs_sorted = docs.sort_values(by='word_count', ascending=False)
```
